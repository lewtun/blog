Over the years I've been asked by former colleagues in academia about how I navigated the transition from science to data science. In the spirit of DRY (Don't Repeat Yourself), I thought I'd collect a few of the things I wish I'd known about when I started out.

> Warning: Everything in this post is based on my personal experience in a specific country (Switzerland) at a specific point in time. Your mileage may vary. I am assuming that you are coming from a hard science background like physics, but maybe little code.

Here was my profile:

* I'd written some C++ code
* Mostly pen and paper
* Did Andrew Ng's classic ML course
* Read stats books
* Had no idea what data science really was.

## Understand the data science landscape
_Data science is an overloaded term. Know what you're looking for._

* Talk about huge variation in job descriptions. Distinction between research jobs / sexy things in media vs mundane/ Build a quadrant?
* Understand what areas are interesting and practical: NLP, blah.
* Deep learning is cool, but most value generated by conventional ML (sklearn)
* Sexiest job of the century, breathless articles from Uber, Google etc bias the landscape

## Don't quit your day job

* Talk about trouble with finding a job unemployed
* Use your time to pick up technical skills in a research project => much more compelling
* Are you sure you cannot pursue ML in your research field? I might have stayed in academia if I'd seen what people were doing with ML!
